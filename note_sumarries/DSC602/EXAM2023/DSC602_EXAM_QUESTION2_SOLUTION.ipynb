## Part 1: Import Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Set style for better-looking plots
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## Part 2: Create the Data Frame

Creating a pandas DataFrame from the provided table with sales data for different sales representatives and product categories.

```python
# Create the data dictionary from the provided table
data = {
    'Sales_Name': ['Henry', 'Peter', 'Nichoulas', 'Mary', 'Dan', 'Oussy', 'Maboukam'],
    'Land_sales': [12, 15, 25, 8, 5, 0, 10],
    'Clothe_sales': [2500, 1900, 1800, 5000, 2850, 8005, 5000],
    'Book_sales': [2300, 2000, 180, 5000, 20, 12, 15]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("Sales Data DataFrame:")
print(df)
print("\nDataFrame Info:")
print(df.info())
print("\nBasic Statistics:")
print(df.describe())
```

## Part 3.1: Bar Plot of Sales Volume

**Theoretical Explanation:**
A bar plot (or bar chart) is one of the most fundamental visualization idioms for comparing values across categories. It uses rectangular bars with lengths proportional to the values they represent. Bar plots are effective for:
- Comparing values across categories (sales representatives in this case)
- Showing differences between groups
- Displaying discrete, quantitative data
- Easy value reading and comparison

In this visualization, we're comparing total sales volume (sum of all product types) for each sales representative.

```python
# Calculate total sales volume for each person
total_sales = df['Land_sales'] + df['Clothe_sales'] + df['Book_sales']
df['Total_Sales'] = total_sales

# Create bar plot
plt.figure(figsize=(12, 6))
bars = plt.bar(df['Sales_Name'], df['Total_Sales'], color='steelblue', edgecolor='navy', alpha=0.7)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
            f'{int(height)}',
            ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.xlabel('Sales Representative', fontsize=12, fontweight='bold')
plt.ylabel('Total Sales Volume', fontsize=12, fontweight='bold')
plt.title('Total Sales Volume by Sales Representative', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("\nTotal Sales Volume by Representative:")
print(df[['Sales_Name', 'Total_Sales']].sort_values('Total_Sales', ascending=False))
```

## Part 3.2: Pie Chart of Item Sales

**Theoretical Explanation:**
A pie chart displays data as slices of a circle, where each slice represents a proportion of the whole. Pie charts are effective for:
- Showing part-to-whole relationships
- Displaying categorical data as percentages
- Visualizing composition (how different parts make up a total)
- Quick understanding of relative proportions

In this visualization, we're showing what proportion of total sales comes from each product category (Land, Clothes, Books).

```python
# Calculate total sales by category
category_totals = {
    'Land Sales': df['Land_sales'].sum(),
    'Clothing Sales': df['Clothe_sales'].sum(),
    'Book Sales': df['Book_sales'].sum()
}

# Create pie chart
plt.figure(figsize=(10, 8))
colors = ['#ff9999', '#66b3ff', '#99ff99']
explode = (0.05, 0.05, 0.05)  # Slightly separate all slices

wedges, texts, autotexts = plt.pie(category_totals.values(), 
                                     labels=category_totals.keys(),
                                     autopct='%1.1f%%',
                                     startangle=90,
                                     colors=colors,
                                     explode=explode,
                                     shadow=True,
                                     textprops={'fontsize': 11, 'fontweight': 'bold'})

# Enhance appearance
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(10)
    autotext.set_fontweight('bold')

plt.title('Proportion of Sales by Product Category', fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.show()

print("\nTotal Sales by Product Category:")
for category, total in category_totals.items():
    percentage = (total / sum(category_totals.values())) * 100
    print(f"{category}: {total} ({percentage:.1f}%)")
```

## Part 3.3: Box Plot of Item Sales

**Theoretical Explanation:**
A box plot (or box-and-whisker plot) displays the distribution of data using quartiles:
- **Box**: Shows the middle 50% of data (interquartile range, IQR)
- **Line in box**: Represents the median (50th percentile)
- **Whiskers**: Extend to show the range of data
- **Points**: Outliers (data points beyond 1.5 × IQR)

Box plots are excellent for:
- Comparing distributions across categories
- Identifying outliers
- Understanding data spread and skewness
- Detecting inconsistencies in data

In this visualization, we're comparing the distribution of sales across the three product categories.

```python
# Prepare data for box plot
box_data = [
    df['Land_sales'],
    df['Clothe_sales'],
    df['Book_sales']
]

# Create box plot
plt.figure(figsize=(10, 6))
bp = plt.boxplot(box_data,
                  labels=['Land Sales', 'Clothing Sales', 'Book Sales'],
                  patch_artist=True,
                  notch=False,
                  showmeans=True,
                  meanprops=dict(marker='D', markerfacecolor='red', markersize=8, label='Mean'))

# Customize colors
colors = ['#ff9999', '#66b3ff', '#99ff99']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

plt.ylabel('Sales Volume', fontsize=12, fontweight='bold')
plt.title('Distribution of Sales by Product Category', fontsize=14, fontweight='bold')
plt.grid(axis='y', alpha=0.3)
plt.legend(['Mean'], loc='upper right')
plt.tight_layout()
plt.show()

# Print statistical summary
print("\nStatistical Summary for Each Product Category:\n")
categories = ['Land_sales', 'Clothe_sales', 'Book_sales']
category_names = ['Land Sales', 'Clothing Sales', 'Book Sales']

for cat, cat_name in zip(categories, category_names):
    data = df[cat]
    print(f"\n{cat_name}:")
    print(f"  Min: {data.min()}")
    print(f"  Q1 (25%): {data.quantile(0.25)}")
    print(f"  Median (Q2): {data.median()}")
    print(f"  Q3 (75%): {data.quantile(0.75)}")
    print(f"  Max: {data.max()}")
    print(f"  Mean: {data.mean():.2f}")
    print(f"  Std Dev: {data.std():.2f}")
    print(f"  IQR: {data.quantile(0.75) - data.quantile(0.25)}")
```

## Part 3.4: Area Plot of Item Sales

**Theoretical Explanation:**
An area plot is similar to a line plot but with the area below the lines filled with color. Area plots are effective for:
- Showing trends over categories (especially when ordered)
- Displaying cumulative totals or compositions
- Emphasizing magnitude and change
- Comparing multiple series that should be stacked

In this visualization, we're showing how sales for each product category vary across different sales representatives, with the area filled to emphasize the volume.

```python
# Create area plot (stacked)
plt.figure(figsize=(12, 6))

# Prepare data
x = np.arange(len(df['Sales_Name']))
width = 0.6

# Create stacked area plot
plt.stackplot(x,
              df['Land_sales'],
              df['Clothe_sales'],
              df['Book_sales'],
              labels=['Land Sales', 'Clothing Sales', 'Book Sales'],
              colors=['#ff9999', '#66b3ff', '#99ff99'],
              alpha=0.7)

plt.xlabel('Sales Representative', fontsize=12, fontweight='bold')
plt.ylabel('Sales Volume', fontsize=12, fontweight='bold')
plt.title('Stacked Area Plot: Sales Distribution by Representative and Category', fontsize=14, fontweight='bold')
plt.xticks(x, df['Sales_Name'], rotation=45, ha='right')
plt.legend(loc='upper left', fontsize=10)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Alternative: Non-stacked area plot
plt.figure(figsize=(12, 6))

plt.plot(df['Sales_Name'], df['Land_sales'], marker='o', linewidth=2, label='Land Sales', color='#ff9999')
plt.fill_between(range(len(df['Sales_Name'])), df['Land_sales'], alpha=0.3, color='#ff9999')

plt.plot(df['Sales_Name'], df['Clothe_sales'], marker='s', linewidth=2, label='Clothing Sales', color='#66b3ff')
plt.fill_between(range(len(df['Sales_Name'])), df['Clothe_sales'], alpha=0.3, color='#66b3ff')

plt.plot(df['Sales_Name'], df['Book_sales'], marker='^', linewidth=2, label='Book Sales', color='#99ff99')
plt.fill_between(range(len(df['Sales_Name'])), df['Book_sales'], alpha=0.3, color='#99ff99')

plt.xlabel('Sales Representative', fontsize=12, fontweight='bold')
plt.ylabel('Sales Volume', fontsize=12, fontweight='bold')
plt.title('Area Plot: Sales Trends by Representative and Category', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.legend(loc='best', fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
